# As described in the README, my first attempt will be to try and classify this information according to distance/position. 

```python
#import dependencies

import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/training_set_metadata.csv')
```

Again further simplifying the question, I want to only answer this question based on objects captured with the same cadence. In this event, I am only working with "wide fast deep" images. (By confirming the ddf bool is false). 

```python
wfd_data = data.loc[data['ddf'] == 0]
```

```python
wfd_data
```

Next, galactic coordinates are a better indicator of position when considering far off astronomical phenomena, not to mention is also ra/dec provide position, so it is redundant to keep both. I will now remove one coord. system. 

```python
del wfd_data['ra'], wfd_data['decl']
```

```python
wfd_data['target'].unique()
```

```python
wfd_data = wfd_data[['gal_l','gal_b','hostgal_specz','target']]
```

```python
wfd_data.dropna(inplace=True)
```

```python
wfd_data
```

Now to make a rough model. Can you determine the type of source based on the position and redshift of that object? Let's find out.

```python
y = wfd_data['target']

y = pd.get_dummies(y)

del wfd_data['target']
```

```python
X = wfd_data.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
```

```python
rf = MultiOutputClassifier(RandomForestClassifier())
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

# Outcome: Based on just location and distance, you get a testing accuracy of ~21%, which is better than random ( 1/ 14 shot = 7%), but still terrible. Furthermore there is supposedly a 15th class to consider, and that has not been included here. So in conclusion, this method fails.
